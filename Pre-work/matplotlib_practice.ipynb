```python
# Import Dependencies
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot.
plt.plot(x_axis, y_axis, marker="*", color="blue", linewidth=2, label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Set the y limit between 0 and 45.
plt.ylim(0, 45)
# Create a title.
plt.title("PyBer Fare by Month")
# Add a grid.
plt.grid()
# Add the legend.
plt.legend()
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis, marker = 'd', color ='green', linewidth = 2, label='Boston')
ax.legend()
ax.set_title('PyBer Fare by Month')
ax.set_xlabel('Date')
ax.set_ylabel('Fare($)')
ax.grid()
ax.set_ylim(0,45)
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot.
plt.bar(x_axis, y_axis, color="green", label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
```

```python
# Create the plot.
plt.barh(x_axis, y_axis, color="darkmagenta", label='Boston')
# Create labels for the x and y axes.
plt.xlabel("Fare($)")
plt.ylabel("Date")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
plt.gca().invert_yaxis()
```

```python
plt.plot(x_axis, y_axis, 'o', color='green')
```

```python
plt.scatter(y_axis, x_axis, color='red', label='Chicago')
plt.ylabel('Date')
plt.xlabel('Fare($)')
plt.xlim(0,50)
plt.gca().invert_yaxis()
plt.legend()
```

```python
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
#plt.scatter(x_axis, y_axis, color='red', s=y_axis_larger, label='Chicago')
plt.scatter(x_axis, y_axis, color='red', s=[i*3 for i in y_axis], label='Chicago')
plt.xlabel('Date')
plt.ylabel('Fare($)')
plt.ylim(0,45)
#plt.gca().invert_yaxis()
plt.legend()
```

```python
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, s=[i*5 for i in y_axis], c='skyblue', label="Boston", linewidth=2, alpha=.8, edgecolors='k')
ax.invert_yaxis()
ax.set_ylabel('Date')
ax.set_xlabel('Fare($)')
ax.set_title('PyBer Fare by Month')
ax.set_xlim(0,50)
ax.legend()
```

```python
plt.subplots(figsize=(8,8))
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
plt.pie(y_axis, labels=x_axis, explode = explode_values, colors=colors, autopct="%.1f%%")

plt.show()
```
